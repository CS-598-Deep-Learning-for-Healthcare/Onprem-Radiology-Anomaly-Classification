```python
test_tables = [
    "workspace.default.mimic_cxr_test_set_label_explanation_extract_databricks_qwen3_next_80b_a3b_instruct_v3",
    "workspace.default.mimic_cxr_test_set_label_explanation_extract_gpt_5_1_v2",
    "workspace.default.mimic_cxr_test_set_label_explanation_extract_llama_4_maverick_v1",
]

test_tables
```

```python
for tbl in test_tables:
    print(f"Adding confidence column to {tbl}")
    
    spark.sql(f"""
        CREATE OR REPLACE TABLE {tbl} AS
        WITH base AS (
          SELECT
            subject_id,
            study_id,
            findings,
            impression,
            label,
            explanation,
            raw_ai_output,
            CONCAT(
              COALESCE(findings, ''),
              ' ',
              COALESCE(impression, '')
            ) AS context_text
          FROM {tbl}
        )
        SELECT
          subject_id,
          study_id,
          findings,
          impression,
          label,
          explanation,
          CASE
            WHEN explanation IS NULL
              OR TRIM(explanation) = ''
            THEN NULL
            ELSE ai_similarity(context_text, explanation)
          END AS confidence,
          raw_ai_output
        FROM base
    """)
    
    print(f"Done: {tbl}\n")

```
